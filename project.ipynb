## 1. Спектральная плотность

Ковариационная функция $R_{\varepsilon}(\tau)$

Для нахождения ковариационной функции $R_{\varepsilon}(\tau)$ воспользуемся обратным преобразованием Фурье:
$\large
R_{\varepsilon}(\tau) = \int_{-\infty}^{-7}0 \cdot e^{2 \pi i \lambda \tau}d \lambda + \int_{-7}^{7} 1 \cdot e^{2 \pi i \lambda \tau}d \lambda + \int_{7}^{+ \infty} 0 \cdot e^{2 \pi i \lambda \tau}d \lambda = \frac{e^{14 \pi i \tau}-e^{-14 \pi i \tau}}{2\pi i \tau}=\frac{\sin(14\pi\tau)}{\pi \tau}
$
$$ R_{\varepsilon}(\tau) = \frac{\sin(14\pi\tau)}{\pi \tau}$$

Ковариационная функция $R_k(\tau)$

Из свойств пуассоновского потока:
$k(s)-k(t) \sim \Pi(\sigma)$ для произвольных моментов $t,s (0 \leq t < s)$, где $\sigma=\lambda|t-s|$ : $p\{X=m\}=\frac{\sigma^m e^{-\sigma}}{m!}, m=0,1,... $

Для фиксированного события $H_k(t)=\{t\in[\tau_k,\tau_{k+1}] \},t\geq 0,k=0,1,...$ образуют систему вероятностных гипотез.

Применим формулу полного математического ожидания:

$
M[x(t)]=\sum\limits_{k = 0}^{\infty}M[x(t)|H_k(t)]P\{H_k(t)\}=
\sum\limits_{k = 0}^{\infty}M[V_k|H_k(t)]P\{H_k(t)\}=
\sum\limits_{k = 0}^{\infty}M[V_k]P\{H_k(t)\}=0
$

Для заданных $0 \leq s <t $ события $\{H_k(s)H_i(t),k,i=0,1,...\}$ - такие системы вероятностных гипотез, поэтому

$cov\{x(t),x(s)\}=M[x(t),x(s)]=
\sum\limits_{k = 0}^{\infty}\sum\limits_{i = 0}^{\infty}M[x(t),x(s)|H_k(s)H_i(t)]P\{H_k(s)H_i(t)\}$

$
k=i:\\
M[x(t),x(s)|H_k(s)H_i(t)]=
M[V_k^2|H_k(s)H_k(t)]=
M[(V_k-M[V_k])^2]=
D[V_k]=d=6
$

$
k\neq i:\\
M[(V_k(t)-M[V_k(t)])(V_i(s)-M[V_i(s)])|H_k(s)H_i(t)]=
M[(V_k(t)-M[V_k(t)])(V_i(s)-M[V_i(s)])]=0
$

Тогда $Cov(x(t),x(s))=D[P\{H\}]$, где $H$ имеет вид:

$
H=\cup_{k=0}^{\infty}H_k(s)H_k(t)=\{\tau_k \notin[s,t] , k \ngeq 1 \} =\{X=0\}
$

Следовательно: $cov\{x(t),x(s)\}=de^{-\lambda|t-s|}=R_x(t-s).$
Таким образом, $x(t)$ - ССФ с ковариационной функцией $R_x(τ) = d\ e^{-λ_0|τ|}$.

$$
R_k(\tau)=6\cdot e^{-|\tau|}
$$


Пусть $F(k(t),\xi(t),\varepsilon(t)) = k(t)\cdot\xi^2(t)+\varepsilon(t)$

Найдем ковариационную функцию  произведения $p(t)=k(t)\cdot\xi^2(t)$:

так как процессы независимы и центрированы по условию, то $M[k(t)\cdot \xi^2(t)] = M[k(t)]\cdot M[\xi^2(t)] = 0$

так как процесс $\xi(t)$ - гауссовский по условию, то $M[X^2_i, X^2_j] = D[X_i] \cdot D[X_j] + 2\cdot Cov(X_i, X_j)$

$R_p(t, s)=Cov(k(t)\cdot \xi^2(t), k(s)\cdot \xi^2(s)) = M[k(t)\cdot \xi^2(t) \cdot k(s)\cdot \xi^2(s)] - M[k(t)\cdot \xi^2(t)]\cdot M[k(t) \cdot \xi^2(t)] = M[k(t)\cdot k(s)]\cdot M[\xi^2(t) \cdot \xi^2(s)] = R_{k}(\tau) \cdot (D[\xi(t)]\cdot D[\xi(s)] + 2 \cdot R_{\xi}(\tau))$

$$=6 \cdot e^{-|\tau|} \cdot (25 + 10e^{-3|\tau|}cos(8\tau))$$

Найдем ковариационную функцию $F(k(t),\xi(t),\varepsilon(t))$

так как процессы независимы и центрированы по условию

$R_F(\tau)=Cov(p(t)+\varepsilon(t), p(s)+\varepsilon(s)) = M[(p(t)+\varepsilon(t))\cdot (p(s)+\varepsilon(s))] - M[p(t) + \varepsilon(t)]\cdot M[p(s) + \varepsilon(s)] = M[p(t)\ p(s) + p(t)\varepsilon(s) + p(s)\varepsilon(t) + \varepsilon(t) \varepsilon(s)]= M[p(t)\cdot p(s)] + M[\varepsilon(t) \varepsilon(s)] = R_{p}(\tau) + R_{\varepsilon}(\tau)$

$$R_{F}=6 \cdot e^{-|\tau|} \cdot (25 + 10e^{-3|\tau|}cos(8\tau))+\frac{\sin(14\pi\tau)}{\pi \tau}.$$

Для нахождения спектральной плотности случайного процесса $F(k(t),\xi(t),\varepsilon(t))$ воспользуемся преобразованием Фурье ковариационной функции этого процесса:
$$S_{F}(\lambda) = \frac{1}{2\pi} \int_{-\infty}^{\infty} R_{F}(\tau) \cdot e^{-i\tau \lambda}\; d\tau.$$


```python
import numpy as np
import sympy as sp
import scipy as scp

tau, λ = sp.symbols('tau λ')
R_ksi = 5 * sp.exp(-3 * abs(tau)) * sp.cos(8*tau)
S_eps = sp.Heaviside(7 - abs(λ), 0)

lambda_p = 1
d_p = 6
```

```python
R_k = d_p * sp.exp(-lambda_p * abs(tau))

R_F_no_eps = R_k * ((R_ksi.subs({tau:0}))**2 + 2 * R_ksi)

R_eps = sp.sin(14*sp.pi*tau) / (sp.pi * tau)

R_F = R_F_no_eps + R_eps
R_F
```

```python
S_F = sp.fourier_transform(1/(2*sp.pi)*R_F, tau, λ/(2*sp.pi))
S_F
```

##### Спектральная плотность для процесса $\eta(t+10)$

$H(λ)\ e^{i\ λ\ (τ\ +\ 10)} = e^{i\ λ\ τ}; \\
H(λ) = e^{-10i\ λ}. \\
Отсюда\ получаем,\ что\ |H(λ)|^2 = |e^{-10i\ λ}|^2 = 1^2.$

$S_η(λ) = |H(λ)|^2 S_F(λ)$

```python
H_λ = 1
S_eta = H_λ * S_F
S_eta
```

## Ковариационная функция

Ковариационная функция для выходного сигнала

Ковариационная функция $\eta(t+10)$ выражается с помощью обратного преобразования Фурье
$$
\begin{align}
R_η(\tau) = \int_{-\infty}^{\infty} S_η(τ)\ e^{i τ λ} dλ
\end{align}
$$

```python
R_eta = R_F
R_eta
```

## Дисперсия

Дисперсия для выходного сигнала

Дисперсия выражается как значение ковариационной функции в нуле (с учетом предельной точки в $\lim\limits_{\tau \to 0} \frac{\sin(14\pi\tau)}{\pi \tau} = 14$).

Посчитаем дисперсию процесса $η(t+10):$

$$D_η = R_η(0) = 6\cdot (25+10) + 14 = 224.$$

## Доказательство

Доказательство стационарности правой части

Так как $k(t)$ и $\xi(t)$ независимы по исловию, а $k(t)$ и $\varepsilon(t)$ центрированы по условию, то
$M[k(t)\xi^2(t)+\varepsilon(t)]=M[k(t)\xi^2(t)]+M[\varepsilon(t)]=M[k(t)]M[\xi^2(t)]+M[\varepsilon(t)]=0$ т.е. мат. ожидание стационарно.

Дисперсия $D[\eta(t)] = 224$ не зависит от времени т.е. дисперсия стационарна.

Ковариационная функция $R_{F}(\tau)=6 \cdot e^{-|\tau|} \cdot (25 + 10e^{-3|\tau|}cos(8\tau))+\frac{\sin(14\pi\tau)}{\pi \tau}$ зависит только от $\tau$ поэтому она является стацонарной.

Из всего вышеперечисленного делаем вывод, что правая часть представляет собой стационарный процесс.
